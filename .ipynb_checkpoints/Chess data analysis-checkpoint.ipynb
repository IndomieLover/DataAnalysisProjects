# OPTIMAL CHESS POSITION
<img src="./chess.jpeg" width=1000 align="left">

In this kernel, we will investigate about the optimal position for 5 major pieces of chess, which are King, Queen, Rook, Bishop, and Knights.

### Importing libraries needed

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import csv
import seaborn as sns
%matplotlib inline
```

### Read chess games from lichess

```python
chess_raw_df = pd.read_csv('games.csv')
chess_df = chess_raw_df[chess_raw_df['rated']==True]
chess_df = chess_raw_df[chess_raw_df['victory_status']=="mate"]
```

### Initializing heatmap

```python
heatmap={
    "win":{
        "white":{
            "K":np.zeros((8,8),dtype=int), #King
            "Q":np.zeros((8,8),dtype=int), #Queen
            "R":np.zeros((8,8),dtype=int), #Rook
            "B":np.zeros((8,8),dtype=int), #Bishop
            "N":np.zeros((8,8),dtype=int)  #Knight
        },
        "black":{
            "K":np.zeros((8,8),dtype=int), #King
            "Q":np.zeros((8,8),dtype=int), #Queen
            "R":np.zeros((8,8),dtype=int), #Rook
            "B":np.zeros((8,8),dtype=int), #Bishop
            "N":np.zeros((8,8),dtype=int)  #Knight
        }
    },
    "lose":{
        "white":{
            "K":np.zeros((8,8),dtype=int), #King
            "Q":np.zeros((8,8),dtype=int), #Queen
            "R":np.zeros((8,8),dtype=int), #Rook
            "B":np.zeros((8,8),dtype=int), #Bishop
            "N":np.zeros((8,8),dtype=int)  #Knight
        },
        "black":{
            "K":np.zeros((8,8),dtype=int), #King
            "Q":np.zeros((8,8),dtype=int), #Queen
            "R":np.zeros((8,8),dtype=int), #Rook
            "B":np.zeros((8,8),dtype=int), #Bishop
            "N":np.zeros((8,8),dtype=int)  #Knight
        }
    }
}
x_axis_labels = ["a","b","c","d","e","f","g","h"]
y_axis_labels = [1,2,3,4,5,6,7,8]
```

### Adding heatmap per chess pieces

```python
for i in range(len(chess_df)//2):
    splitted=chess_df.iloc[i]['moves'].split()
    win=chess_df.iloc[i]['winner'].split()[0]
    for en, spl in enumerate(splitted):
        if (spl[-2]=='='):
            continue
        if (spl[-1]=='+' or spl[-1]=='#'):
            spl=spl[:-1]
        color = "white" if en%2==0 else "black"
        stat = "win" if win==color else "lose"
        if(spl[0]=='Q' or spl[0]=='K' or spl[0]=='R' or spl[0]=='B' or spl[0]=='N'):
            heatmap[stat][color][spl[0]][int(spl[-1])-1][ord(spl[-2])-ord('a')]+=1
```

## King Position
<img src="./king.png" width=100 align="left">

```python
plt.subplots(figsize=(10,5))
plt.title("White Player's Winning King Position", size=15)
sns.heatmap(heatmap["win"]["white"]["K"],square=True,xticklabels=x_axis_labels,yticklabels=y_axis_labels)
```

```python
plt.subplots(figsize=(10,5))
plt.title("White Player's Losing King Position", size=15)
sns.heatmap(heatmap["lose"]["white"]["K"],square=True,xticklabels=x_axis_labels,yticklabels=y_axis_labels)
```

```python
plt.subplots(figsize=(10,5))
plt.title("Black Player's Winning King Position", size=15)
sns.heatmap(heatmap["win"]["black"]["K"],square=True,xticklabels=x_axis_labels,yticklabels=y_axis_labels)
```

```python
plt.subplots(figsize=(10,5))
plt.title("Black Player's Losing King Position", size=15)
sns.heatmap(heatmap["lose"]["black"]["K"],square=True,xticklabels=x_axis_labels,yticklabels=y_axis_labels)
```

Based on the data above, we can interpret a few conclusions:
* Losing King piece position for both white and black players located nearer to their original row, which are wither 8 or 1 depending on the color
* Generally, winning players moves the king further from their original position
* Most played position for kings regardless their winning condition is around d - h tile
